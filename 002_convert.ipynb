```python
import os
import PyPDF2

def converter_pdfs_para_txt(
    pasta_pdf="data2025/pdf2025",
    pasta_txt="data2025/txt2025"
):
    """
    Converte todos os PDFs da pasta pdf2025 para arquivos TXT em txt2025.
    """

    os.makedirs(pasta_txt, exist_ok=True)

    pdfs = [f for f in os.listdir(pasta_pdf) if f.lower().endswith(".pdf")]

    if not pdfs:
        print("⚠️ Nenhum PDF encontrado em", pasta_pdf)
        return

    for nome_pdf in pdfs:
        caminho_pdf = os.path.join(pasta_pdf, nome_pdf)
        nome_txt = nome_pdf.replace(".pdf", ".txt")
        caminho_txt = os.path.join(pasta_txt, nome_txt)

        if os.path.exists(caminho_txt):
            print(f"⏩ Já existe TXT, pulando: {nome_txt}")
            continue

        print(f"\n📄 Convertendo {nome_pdf}")

        try:
            with open(caminho_pdf, "rb") as f:
                leitor = PyPDF2.PdfReader(f)

                if getattr(leitor, "is_encrypted", False):
                    try:
                        leitor.decrypt("")
                    except Exception:
                        print(f"🔒 PDF criptografado: {nome_pdf}")
                        continue

                texto_total = ""

                for i, pagina in enumerate(leitor.pages):
                    texto = pagina.extract_text() or ""
                    texto_total += f"\n--- Página {i+1} ---\n{texto}"

            with open(caminho_txt, "w", encoding="utf-8") as out:
                out.write(texto_total)

            print("✅ TXT salvo")

        except Exception as e:
            print(f"❌ Erro ao converter {nome_pdf}: {e}")

    print("\n🎉 Finalizado!")

```

```python
import os
import PyPDF2
import hashlib
```

```python
def hash16(texto):
    return hashlib.md5(texto.encode("utf-8")).hexdigest()[:16]
```

```python
def converter_pdfs_para_txt_com_relatorio(anos, base_dir=".", relatorio="relatorio_erros_conversao.txt"):
    """
    Converte PDFs para TXT por ano e gera relatório de erros com LINK e ANO.
    """

    erros = []

    for ano in anos:
        pasta_pdf = f"{base_dir}/data{ano}/pdf{ano}"
        pasta_txt = f"{base_dir}/data{ano}/txt{ano}"
        arquivo_links = f"{base_dir}/ppgsCSA_{ano}_link.txt"

        print(f"\n📅 Processando {ano}")

        os.makedirs(pasta_txt, exist_ok=True)

        # --- mapa hash -> link ---
        mapa_links = {}
        if os.path.exists(arquivo_links):
            with open(arquivo_links, "r", encoding="utf-8") as f:
                for link in f:
                    link = link.strip()
                    if link:
                        mapa_links[hash16(link)] = link
        else:
            print(f"⚠️ Arquivo de links não encontrado: {arquivo_links}")

        if not os.path.exists(pasta_pdf):
            print(f"⚠️ Pasta não encontrada: {pasta_pdf}")
            continue

        for nome_pdf in os.listdir(pasta_pdf):
            if not nome_pdf.lower().endswith(".pdf"):
                continue

            hash_pdf = nome_pdf.replace(".pdf", "")
            caminho_pdf = os.path.join(pasta_pdf, nome_pdf)
            caminho_txt = os.path.join(pasta_txt, hash_pdf + ".txt")

            if os.path.exists(caminho_txt):
                continue

            try:
                with open(caminho_pdf, "rb") as f:
                    leitor = PyPDF2.PdfReader(f)

                    if getattr(leitor, "is_encrypted", False):
                        leitor.decrypt("")

                    texto_total = ""
                    for pagina in leitor.pages:
                        texto_total += pagina.extract_text() or ""

                with open(caminho_txt, "w", encoding="utf-8") as out:
                    out.write(texto_total)

            except Exception as e:
                link = mapa_links.get(hash_pdf, "LINK_NAO_ENCONTRADO")
                erros.append(f"Ano:{ano}\tLink:{link}\tErro:{str(e)}")

    # --- salvar relatório ---
    with open(relatorio, "w", encoding="utf-8") as f:
        for linha in erros:
            f.write(linha + "\n")

    print(f"\n📄 Relatório gerado: {relatorio}")
```

```python
anos = [2020, 2021, 2022, 2023, 2024, 2025]

converter_pdfs_para_txt_com_relatorio(
    anos=anos,
    base_dir=".",
    relatorio="relatorio_erros_conversao.txt"
)
```
